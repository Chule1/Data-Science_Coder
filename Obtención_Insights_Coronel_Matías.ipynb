```python
yes_count = pd.DataFrame(df_BM['y'].value_counts()).reset_index(names='Yes Value').rename(columns={'y':'Count'})
yes_count['Percentage']=(100*yes_count['Count']/(yes_count['Count'].sum())).round(3)
yes_count
```

```python
fig,ax=plt.subplots(figsize=(5,5))
ax.pie(yes_count['Count'],
       labels=[f'{y}\n{p:.2f}% - ({c})' for y,p, c in zip(yes_count['Yes Value'],yes_count['Percentage'], yes_count['Count'])],
       startangle=0,
       colors=[sns.color_palette("pastel")[0],sns.color_palette("pastel")[1]])
```

### **Influencia de la edad y estado civil**

```python
age_beh=pd.merge(df_BM_yes.groupby('age')['y'].count().reset_index(),
                 df_BM_no.groupby('age')['y'].count().reset_index(),
                 on='age', how='outer').fillna(0)
age_beh=age_beh.rename(columns={"y_x": "yes", "y_y": "no"})
age_beh['total']=age_beh['yes']+age_beh['no']
age_beh['prop_yes']=age_beh['yes']/age_beh['total']
age_beh['prop_no']=age_beh['no']/age_beh['total']
age_beh['porc_yes']=(100*age_beh['prop_yes']).round(2)
age_beh['porc_no']=(100*age_beh['prop_no']).round(2)
age_beh.sample(10)
```

```python
colores_seaborn = sns.color_palette("pastel")
lista_colores = [f'rgb({int(color[0] * 255)}, {int(color[1] * 255)}, {int(color[2] * 255)})' for color in colores_seaborn]

fig = px.bar(age_beh, x='age', y=['porc_yes', 'porc_no'], color_discrete_map=dict(zip(['porc_no', 'porc_yes'], lista_colores)),
             title='Porcentaje de aceptación o no por edad',
             labels={'age': 'Edad', 'value': 'Porcentaje', 'variable': 'Resultado'})

#Dandole formato al eje X
fig.update_xaxes(tickangle=-90, tickmode='array', tickvals=age_beh['age'])

fig.update_layout(plot_bgcolor='rgba(0,0,0,0)', paper_bgcolor='rgba(0,0,0,0)',
                  title={'text': 'Porcentaje de aceptación o no - Por edad',
                         'y': 0.85, 'x': 0.5, 'xanchor': 'center', 'yanchor': 'top'},
                  title_font=dict(size=18, color='black'),
                  xaxis=dict(linecolor='grey', showgrid=False, gridcolor='grey',gridwidth=1, griddash='dot'),
                  yaxis=dict(linecolor='grey', showgrid=False, gridcolor='grey',gridwidth=1, griddash='dot'))
fig.show()
```

<p align='justify'> En el gráfico anterior se puede observar como, en proporciones, el segmento etario 60-90 presenta niveles de aceptación del producto en el entorno, aproximado, de 30-80%. Estos valores son muy superiores a los que se ven en las edades 20-60 donde no se llega a lograr el 20% de éxito en las llamadas.
<br> Recordando lo visto en la frecuencia absoluta de llamadas a cada grupo etario, 20-60 es el mas cuantioso pero con menor porcentaje de éxito, mientras que 60-90 tiene la menor cantidad de llamadas pero el más alto índice de éxito. <br> Esto podría poner en evaluación la estrategia de marketing a seguir, si se decide dirigir los esfuerzos a aumentar el porcentaje de éxito en el grupo etario mas cuantioso o si se prefiere trabajar en aumentar los clientes de edad más avanzada con la hipotesis de que mantendrán los niveles de aceptación vistos en este gráfico.   </p>

### **Detección de outliers**

<p align='justify'>Para el analizar si existen outliers, teniendo variables categóricas que se han convertido en numéricas se podrían utilizar propios de este último tipo de variables. Sin embargo, la mayoría de las variables del dataset con el que se trabaja son nominales (no tienen una jerarquía definida) y habiendo usado One Hot Encoding la distancia entre puntos no tiene un significado real (siendo esto la base de los métodos de detección de outliers en variables numéricas) </p>

<p align='justify'> A continuación se intentarán detectar los outliers utilizando un método de aprendizaje automático diseñado para la detección de anomalías en datos categóricos. Este algoritmo se llama "Isolation Forest" </p>

<p align='justify'> El Isolation Forest es un algoritmo de detección de anomalías basado en árboles. Su enfoque se centra en aislar las instancias anómalas mediante la construcción de árboles de decisión de manera aleatoria.</p>

<p align='justify'> A diferencia de los métodos tradicionales que buscan construir regiones densas alrededor de las instancias normales, el Isolation Forest sigue un enfoque inverso. Selecciona aleatoriamente una característica y un valor umbral y construye el árbol hasta que cada instancia se aísle en sus propias hojas. Las instancias anómalas tienden a requerir menos divisiones para ser aisladas, y por lo tanto, se encuentran más arriba en los árboles.</p>

<p align='justify'> Las instancias que se encuentran cerca de la raíz del árbol (requieren menos divisiones) se consideran más probables de ser anómalas. La puntuación de anomalía se calcula basándose en la profundidad del nodo en el que se encuentra la instancia. </p>

```python
from sklearn.ensemble import IsolationForest
```

### **Forward Selection**

```python
def forward_selection(data, target, significance_level=0.01):
    """
    params
      -data: Pandas DataFrame con las características
      -target: Pandas Series con la variable target
      -significance_level: Nivel de significancia para evaluar los coeficientes de las caracteríticas.
    return
      -best_features: lista con las características seleccionadas como mejores
    """
    initial_features = data.columns.tolist()
    best_features = []
    while (len(initial_features)>0):
        remaining_features = list(set(initial_features)-set(best_features))
        new_pval = pd.Series(index=remaining_features)
        for new_column in remaining_features:
            model = sm.OLS(target, sm.add_constant(data[best_features+[new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        min_p_value = new_pval.min()
        if(min_p_value<significance_level):
            best_features.append(new_pval.idxmin())
        else:
            break
    return best_features
```

```python
forward_best_features = forward_selection(X,Y,significance_level = 0.05)
forward_best_features,len(forward_best_features)
```

```python
def backward_elimination(data, target,significance_level = 0.05):
    """
    params
      -data: Pandas DataFrame con las características
      -target: Pandas Series con la variable target
      -significance_level: Nivel de significancia para evaluar los coeficientes de las caracteríticas.
    return
      -best_features: lista con las características seleccionadas como mejores
    """
    features = data.columns.tolist()
    while(len(features)>0):
        features_with_constant = sm.add_constant(data[features])
        p_values = sm.OLS(target, features_with_constant).fit().pvalues[1:]
        max_p_value = p_values.max()
        if(max_p_value >= significance_level):
            excluded_feature = p_values.idxmax()
            features.remove(excluded_feature)
        else:
            break
    return features
```

```python
backward_best_features = backward_elimination(X,Y,significance_level = 0.05)
backward_best_features,len(backward_best_features)
```

```python
def stepwise_selection(data, target,SL_in=0.05,SL_out = 0.05):
    initial_features = data.columns.tolist()
    best_features = []
    while (len(initial_features)>0):
        remaining_features = list(set(initial_features)-set(best_features))
        new_pval = pd.Series(index=remaining_features)
        for new_column in remaining_features:
            model = sm.OLS(target, sm.add_constant(data[best_features+[new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        min_p_value = new_pval.min()
        if(min_p_value<SL_in):
            best_features.append(new_pval.idxmin())
            while(len(best_features)>0):
                best_features_with_constant = sm.add_constant(data[best_features])
                p_values = sm.OLS(target, best_features_with_constant).fit().pvalues[1:]
                max_p_value = p_values.max()
                if(max_p_value >= SL_out):
                    excluded_feature = p_values.idxmax()
                    best_features.remove(excluded_feature)
                else:
                    break
        else:
            break
    return best_features
```

```python
stepwise_best_features = stepwise_selection(X,Y,SL_in=0.05,SL_out = 0.05)
stepwise_best_features,len(stepwise_best_features)
```

### **Tratamiento del desbalance de clases**

<p align='justify'> Como se ha comentado anteriormente, que en el conjunto de datos existan mucha más cantidad de registros con 'y'=0 que con 'y'=1 podría estar afectando el entrenamiento del modelo. Esto se conoce como desequilibrio o desbalance de clases.</p>

<p align='justify'> Existen varios métodos y técnicas específicamente diseñados para abordar el problema del desequilibrio de clases en conjuntos de datos. Algunos de los métodos de aprendizaje desbalanceado más comunes son: </p>

<ol> <li> Sobremuestreo de la Clase Minoritaria:

<ul> <li> <p align='justify'> Método de Duplicación (Oversampling): Aumentar el número de instancias de la clase minoritaria mediante la duplicación de las muestras existentes o la generación de nuevas muestras sintéticas.</p>

<li><p align='justify'>SMOTE (Synthetic Minority Over-sampling Technique): Generar muestras sintéticas para la clase minoritaria creando combinaciones lineales entre instancias vecinas.</p> </ul>

<li>Submuestreo de la Clase Mayoritaria:

<ul> <li> <p align='justify'>Eliminación Aleatoria (Undersampling): Reducir el número de instancias de la clase mayoritaria al azar para equilibrar las clases.</p>
<li><p align='justify'>Tomek Links: Eliminar instancias cercanas de la clase mayoritaria que forman enlaces de Tomek con instancias de la clase minoritaria.</p>
</ul></ol>

</p>

```python
from imblearn.over_sampling import SMOTE
```

```python
#Sobremuestreo - SMOTE
smote = SMOTE(random_state=42)
X_train_resampled, Y_train_resampled = smote.fit_resample(X_train, Y_train)
RF_model_smote = RandomForestClassifier(max_depth=None,n_estimators=50, random_state=42, min_samples_split=10, min_samples_leaf=1)
RF_model_smote.fit(X_train_resampled, Y_train_resampled)
y_rf_pred_smote=RF_model_smote.predict(X_test)
```

```python
cm_smote = confusion_matrix(Y_test, y_rf_pred_smote)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_smote)
disp.plot()
```

```python
print(classification_report(Y_test, y_rf_pred_smote))
```

<p align='justify'> Como puede observarse, realizando la técnica de sobremuestreo de clase minoritaria, las métricas recall y f1_score para los valores positivos del target mejoraron casi al doble del valor antes obtenido. También se observa una leve mejora en el accuracy ponderado.</p> <br> <p align='justify'>Esto muestra que las técnicas pensadas para tratar el desequilibrio de casos pueden mejorar la performance del modelo, podría tratarse nuevos enfoques, utilizando SMOTE + alguna técnica de Undersampling para analizar si se mejoran aún más. </p>
